```python
import matplotlib.pyplot as plt
import numpy as np
import generate_toy as gt
import seaborn as sns
import pandas as pd
from numpy.random import default_rng

global_rng = default_rng(seed=0)

figsize = (12, 6)
color_palette = sns.color_palette("Dark2")

def plot_examples(func, params, lims, step, noise=0, rng=global_rng, scale=False):
    plt.figure(figsize=figsize)
    for i in range(len(params)):
        X = np.arange(lims[i][0], lims[i][1], step)
        if 0 in X:
            print(X)
        y = func(X, *params[i])
        if scale:
            y = 10 * y/(max([abs(k) for k in y]))
        y_noisy = y + gt.gaussian_noise(y, rng, noise)

        plt.scatter(X, y_noisy, color=color_palette[i], label=f'Example {i} : {params[i]}')
    plt.legend()
```

# Polynomial equation

```python
step = .2
example_parameters = [[2, -2, 0, 0],[0, 2, 2, 0],[0, 0, 2, 2], [2, 0, 0, 2]]

X_limits = [[-2, 2],
            [-2, 2],
            [-2, 2],
            [-2, 2]]

plot_examples(gt.func_poly3, example_parameters, X_limits, step, noise=0)
```

```python
step = .05
example_parameters = [[2, -2, 2, 2],
                      [2, -2, 2, 2],
                      [2, -2, 2, 2],
                      [2, -2, 2, 2]]

X_limits = [[-2, -1],
            [-1, 0],
            [0, 1],
            [1, 2]]

plot_examples(gt.func_poly3, example_parameters, X_limits, step)
```
