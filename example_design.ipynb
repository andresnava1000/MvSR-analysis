```python
import matplotlib.pyplot as plt
import numpy as np
import generate_toy as gt
import seaborn as sns


figsize = (12, 6)
color_palette = sns.color_palette("Dark2")

def plot_examples(func, params, lims, step):
    plt.figure(figsize=figsize)
    for i in range(len(params)):
        X = np.arange(lims[i][0], lims[i][1], step)
        if 0 in X:
            print(X)
        y = func(X, *params[i])
        plt.scatter(X, y, color=color_palette[i], label=f'Example {i} : {params[i]}')
    plt.legend()
```
